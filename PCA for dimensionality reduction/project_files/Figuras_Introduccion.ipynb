# Situaci√≥n problema

```python
#Librerias
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
```

```python
#Generaci√≥n de una base de datos con valores aleatorios
rng = np.random.RandomState(1)
mat= np.dot(rng.rand(3, 3), rng.randn(3, 200)).T

#Generaci√≥n de la primera figura
fig=plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.scatter(mat[:, 0], mat[:, 1],mat[:,2])
plt.show()
```

#Reducci√≥n dimensional

```python
#Reducci√≥n a dos dimensiones
pca2=PCA(n_components=2)
pca2.fit(mat)
mat_2=pca2.transform(mat)
pca1=PCA(n_components=1)
pca1.fit(mat)
mat_1=pca1.transform(mat)
```

```python
print('Dimensiones originales: ', mat.shape)
print('Reducci√≥n a dos dimensiones: ', mat_2.shape)
print('Reducci√≥n a una dimensi√≥n: ',mat_1.shape)
```

```python
#Realizamos una transformaci√≥n inversa para visualizaci√≥n
mat_2i=pca2.inverse_transform(mat_2)
mat_1i=pca1.inverse_transform(mat_1)
```

```python
#Visualizaci√≥n de la reducci√≥n en 3d
fig=plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.scatter(mat[:, 0], mat[:, 1],mat[:,2])
ax.scatter(mat_2i[:, 0], mat_2i[:, 1],mat_2i[:,2])
ax.scatter(mat_1i[:,0],mat_1i[:,1],mat_1i[:,2])
plt.show()
```

```python
#Visualizaci√≥n de la reducci√≥n en 2d
plt.scatter(mat[:,0],mat[:,1])
plt.scatter(mat_2[:,0],mat_2[:,1])
plt.scatter(mat_1i[:,0],mat_1i[:,1])
```
